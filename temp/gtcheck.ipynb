비디오별 GT 슬라이스 길이 정렬도 맞는지 (중간에 누락/추가/순서 꼬임으로 ptr이 밀려 gt끝을 넘어가진 않았는지)

```python
import numpy as np

gt = np.load("list/gt-ucf-RTFM.npy", allow_pickle=True)
nalist = np.load("list/nalist_test_i3d.npy")

ptr = 0
ok = True
for i,(a,b) in enumerate(nalist):
    T = int(b-a)
    gt_i = gt[ptr:ptr+T*16]
    if len(gt_i) != T*16:
        print("Mismatch at video", i)
        ok = False
        break
    ptr += T*16
assert ptr == len(gt)

print("all slices ok?", ok, "final ptr", ptr, "len(gt)", len(gt))
```

비디오별 gt 합 sanity check (중간에 T가 다른 비디오가 잇는지. 전체 T개수는 동일한데 비디오 별 T_i 개수가 다른지 확인)

```python
import numpy as np

gt = np.load("list/gt-ucf-RTFM.npy", allow_pickle=True)
nalist = np.load("list/nalist_test_i3d.npy")

with open("list/ucf-i3d_test_fixed_local.list", "r") as f:
    names = [line.strip().split("/")[-1] for line in f if line.strip()]

ptr = 0
for i in range(20):
    T = int(nalist[i,1] - nalist[i,0])
    gt_i = gt[ptr:ptr+T*16]
    ptr += T*16
    print(i, names[i], "GT_sum=", int(gt_i.sum()))
```
